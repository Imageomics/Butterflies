```python
data = pd.read_csv('/content/drive/MyDrive/TableS2HoyalCuthilletal2019-Kozak_curated.csv', encoding= 'unicode_escape')
hybrid_data = pd.DataFrame(data.loc[data['Hybrid status of additional taxonomic information'] == 'hybrid'])
hybrid_data
```

```python
dataset = pd.concat([hybrid_data]).reset_index(drop=True)
```

```python
filenames = [] 
for i in range(len(dataset)) : 
  filename = dataset['Image filename'][i][:-10] + dataset['Subspecies'][i] + '_'
  if dataset['Species'][i] == 'erato' : 
    filename += 'E.jpg' 
  else : filename += 'M.jpg'
  filenames.append(filename)
```
